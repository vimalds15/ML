```python
import pandas as pd
import numpy as np 
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/linear reg/canada_pci.csv")
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (usd)')
plt.scatter(df.year,df["per capita income"],color="red",marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income'])
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('per capita income',fontsize=20)
plt.scatter(df.year,df["per capita income"],color="red",marker="+")
plt.plot(df.year,reg.predict(df[["year"]]),color="blue")
```
